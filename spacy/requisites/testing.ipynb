```python
import spacy
```

```python
nlp = spacy.load("training.spacy")
doc = nlp("Computer Science 319 or 331; and Computer Science 355 or Computer Engineering 369.")
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
          token.shape_, token.is_alpha, token.is_stop)
```
