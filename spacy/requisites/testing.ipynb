```python
import spacy
```

```python
nlp = spacy.load("./output/model-best")

doc = nlp("Medicine 300, 303, 320, 330, 340, 345, 350, 360, and 370.")
for ent in doc.ents:
    print(ent)
```

```python
# Read lines 
with open("data/prereqs.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        print(line, end="")
        
        doc = nlp(line)
        for ent in doc.ents:
            print("->", ent)
            
        print("\n")
```
