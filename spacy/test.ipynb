```python
import json
import spacy
import random
from spacy import displacy

nlp = spacy.load('../nero_entity_ruler_model')


file = open("testing_data.jsonlines", "r")
lines = file.readlines()

random.shuffle(lines)

for line in lines[:500]:
    line = json.loads(line)
    text = line[0]
    
    doc = nlp(text)
    displacy.render(doc, style="ent")



file.close()

```
