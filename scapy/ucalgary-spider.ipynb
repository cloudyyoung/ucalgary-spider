```python
import pandas as pd
```

```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_sm')
```

```python
ruler = nlp.add_pipe("entity_ruler", config={"validate": True})
patterns = [
    {"label": "PROG", "pattern": "Mathematics", "id": "math"},
    {"label": "PROG", "pattern": "Computer Science", "id": "cpsc"},
    {"label": "PROG", "pattern": "Applied Mathematics", "id": "math"},
]
ruler.add_patterns(patterns)
```

```python
prereq = "Computer Science 319 or 331; and Mathematics 211 or 213; and 3 units from Mathematics 253, 267, 277, 283 or Applied Mathematics 219."
doc = nlp(prereq)
```

```python
data = []
for token in doc:
    data.append([token.text, token.lemma_, token.pos_, spacy.explain(token.pos_), token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop])

pd.DataFrame(data, columns=["Text", "Lemma", "POS", "Explain", "Tag", "Dependent", "Shape", "Alpha", "Stop word"])
```

```python
options = {"compact": True}
displacy.render(doc, style="dep", options=options)
```

```python
displacy.render(doc, style="ent", options=options)
```

```python
data = [(ent.text, ent.label_) for ent in doc.ents]
pd.DataFrame(data, columns=["Text", "Label"])
```
